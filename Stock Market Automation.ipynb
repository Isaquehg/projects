# Importing a Single Quote

```python
from pandas_datareader import data as web
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Quote, data_source, start, end
magalu = web.DataReader('MGLU3.SA', data_source = 'yahoo', start = '01/01/2020', end = '01/01/2021')
display(magalu)
```

```python
#Quotes graphics
magalu['Adj Close'].plot(figsize = (13, 10))#Creating the graphic
plt.show()#Showing it
```

# Stock Portifolio with a XLSX File

```python
companies_df = pd.read_excel('Empresas.xlsx')
display(companies_df)
```

```python
for company in companies_df['Empresas']:
    quote = web.DataReader(f'{company}.SA', data_source='yahoo', start='01/15/2020', end='06/15/2021')
    display(quote)#table
    quote['Adj Close'].plot(figsize = (13, 10))
    plt.show()#graphic
```
