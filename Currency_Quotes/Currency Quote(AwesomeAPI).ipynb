#Picking up Currency Quotes through API from AwesomeAPI

```python
import requests
import json
```

```python
quotes = requests.get("https://economia.awesomeapi.com.br/last/USD-BRL,EUR-BRL,BTC-BRL")#HTTP request
quotes = quotes.json()#converting json dict to python
bitcoin_bid = quotes['BTCBRL']['bid']#iterating through dictionary
print(quotes)
print(f'The bitcoin quotes now is {bitcoin_bid} in Reais')
```
